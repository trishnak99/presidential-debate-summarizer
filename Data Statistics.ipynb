```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from nltk import bigrams
from collections import Counter
```

```python
biden_df = pd.read_csv('data/preprocessed/bidenSentencesPreprocessed.csv')
biden_df_tokens = biden_df["cluster tokens"].tolist()

biden_tokens = " "
for token in biden_df_tokens:
    if type(token) == str:
        biden_tokens += " " + token
        
biden_tokens = biden_tokens.strip().split()
```

```python
my_bigrams = bigrams(biden_tokens)
counts = Counter(my_bigrams)
counts = dict(counts)

dcounts = pd.DataFrame.from_dict(counts, orient='index', columns=['frequency'])
dcounts = dcounts[dcounts.frequency>=4]
dcounts = dcounts.sort_values(by='frequency', ascending=True)
plt.rcParams['figure.figsize'] = [10, 10]
plt.barh(list(map(str, dcounts.index)), dcounts.frequency)
plt.title('Most Frequent Bigrams - Joe Biden')
plt.grid()
plt.show()
```
